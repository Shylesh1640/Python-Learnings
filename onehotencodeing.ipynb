```python
import pandas as pd 
from sklearn.preprocessing import OneHotEncoder
```

```python
data = {'Fruit': ['Apple', 'Banana', 'Orange', 'Apple', 'Orange'],
        'Color': ['Red', 'Yellow', 'Orange', 'Green', 'Orange']}
```

```python
df = pd.DataFrame(data)
```

```python
encoder = OneHotEncoder()
```

```python
encoder_color = encoder.fit_transform(df[['Color']]).toarray()
```

```python
encodee_df = pd.DataFrame(encoder_color, columns=encoder.get_feature_names_out(['Color']))
```

```python
df_encoded = pd.concat([df, encodee_df], axis=1).drop('Color', axis=1)
```

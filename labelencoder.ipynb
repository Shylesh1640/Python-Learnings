```python
import pandas as pd
import sklearn.preprocessing as LabelEncoder
```

```python
data = {'item':['Shirt','Jeans','Dresss','T-shirt','Skirt'],
        'color':['Blue','Black','Red','White','Green'],
        'size':['Medium','Large','Small','Medium','Small']}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
encoder = LabelEncoder
```

```python
df['Encoded_Size'] = encoder.LabelEncoder().fit_transform(df['size'])
```

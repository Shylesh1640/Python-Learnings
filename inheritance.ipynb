# OOPS Inheritance

```python
class Car():
    def __init__(self,windows,doors,engine):
        self.windows = windows
        self.doors = doors
        self.engine = engine
    def drive(self):
        print("person is driving the car")
        
```

```python
car1 = Car(4,4,"V6")
```

```python
car1.windows
```

```python
car1.engine
```

```python
dir(car1)
```

```python
car1.drive()
```

```python
class audi(Car):
    def __init__(self,windows,doors,engine,model):
        super().__init__(windows,doors,engine)
        self.model = model
    def sportmode(self):
        print("audi is in sport mode")
```

```python
car2 = audi(4,4,"V8","Q7")
```

```python
car2.sportmode()
```

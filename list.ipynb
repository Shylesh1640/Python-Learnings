```python
#list Comprehension
```

```python
lst1 = []
def lst_square(lst):
    for i in lst:
        lst1.append(i*i)
    return lst1
```

```python
lst_square([1,2,3,4,5])
```

```python
#using list comprehension
```

```python
lst2 = [1,2,3,4,5]
lst3 = [i*i for i in lst2]
```

```python
print(lst3)
```

```python
[i*i for i in lst2 if i%2==0]
```

```python
[i*i for i in lst2 if i%2!=0]
```

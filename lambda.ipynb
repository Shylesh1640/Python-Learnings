```python
#lambda Function
#anonymous function
#A function that is defined without a name
```

```python
def add(x, y):
    return x + y
```

```python
add(2, 3)
```

```python
add = lambda x, y: x * y
```

```python
add(2, 3)
```

```python
def even(num):
    if num%2==0:
        return True
```

```python
even(4)
```

```python
even1 = lambda num: True if num%2==0 else False
```

```python
even1(45)
```

```python
def add(x,y,z):
    return x + y + z
```

```python
add(2,3,4)
```

```python
add1 = lambda x,y,z: x + y + z
```

```python
add1(2,3,4)
```

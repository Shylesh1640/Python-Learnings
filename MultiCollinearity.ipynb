```python
import pandas as pd 
import statsmodels.api as sm
df_adv = pd.read_csv('Advertising.csv',index_col=0)
```

```python
x = df_adv[['TV','Radio','Newspaper']]
y = df_adv['Sales']
```

```python
df_adv.head()
```

```python
x = sm.add_constant(x)
model = sm.OLS(y,x).fit()
```

```python
model.summary()
```

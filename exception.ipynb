```python
try:
    a=b
except Exception as e:
    print("An error occurred:", e)
```

```python
try:
    a=b
except NameError as ex1:
    print("NameError occurred:", ex1)
except Exception as e:
    print("Some other error occurred:", e)
```

```python
try:
    a=1
    b="hi"
    c=a+b
except TypeError as ex2:
    print("TypeError occurred:", ex2)
```

```python
try:
    a = int(input("Enter a number: "))
    b = int(input("Enter another number: "))
    c = a / b
    d = a * b
    e = a + b
    print("Division:", c)
    print("Multiplication:", d) 
    print("Addition:", e)
except ValueError as ve:
    print("Error: Invalid input. Please enter numeric values.", ve)
except TypeError as te:
    print("Error: Type mismatch encountered.", te)
except ZeroDivisionError as zde:
    print("Error: Cannot divide by zero.", zde)
else:
    print("All operations completed successfully.")
    print("Final Results - Division:", c, "Multiplication:", d, "Addition:", e)
```

```python
#try else and finally
try:
    a = int(input("Enter a number: "))
    b = int(input("Enter another number: "))
    c = a / b
    print("Division:", c)
except ZeroDivisionError as zde:
    print("Error: Cannot divide by zero.", zde)
finally:
    print("Execution completed.")
```

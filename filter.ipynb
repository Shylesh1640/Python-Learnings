```python
import pandas as pd 
```

```python
data = {
    'Age':[25,30,40,22],
    'Income':[30000,45000,60000,25000],
    'Loan Approval':[1,0,1,0]
}
df = pd.DataFrame(data)
df
```

```python
df.corr()
```

```python
corr = df.corr()['Loan Approval']
corr
```

```python
corr = corr.abs()
corr
```

```python
sorted_corr = corr.sort_values(ascending=False)
sorted_corr
```

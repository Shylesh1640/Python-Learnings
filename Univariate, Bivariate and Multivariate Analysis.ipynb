```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#inport iris dataset
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df.shape
```

```python
# univariate Analysis

df_setosa = df.loc[df['species']=='setosa']
```

```python
df_versicolor = df.loc[df['species']=='versicolor']
```

```python
df_verginica = df.loc[df['species']=='virginica']
```

```python
# univariate diagrams

plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']), 'o')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']), 'o')
plt.plot(df_verginica['sepal_length'],np.zeros_like(df_verginica['sepal_length']), 'o')
plt.xlabel('petal_length')
plt.show()
```

```python
#Bivariate Analysis

#take two feartures use seaborn which has inbuilt functions for bivariate analysis

sns.FacetGrid(df,hue='species',height=5).map(plt.scatter,'petal_length','sepal_width').add_legend();
plt.show()
```

```python
#Multivariate Analysis

#pairplot to see pairwise relationship in the dataset

sns.pairplot(df,hue='species',height=3);
```
